```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
mb = open('macbeth', 'r').read()
```

```python
mb_wc = WordCloud(width=800, height=600,background_color='black',
    max_words=100)
```

```python
mb_wc.generate(mb)
plt.figure(figsize=(16,16))
plt.imshow(mb_wc, interpolation='hamming')
plt.axis('off')
plt.show()
```

```python
mb_wc.to_file("McBeth_word_cloud.png")
```
